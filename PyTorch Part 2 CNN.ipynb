```python
class Linear(nn.Module):
    def __init__(self, in_features, out_features, bias=True):
        super(Linear, self).__init__()
        self.in_features=in_features
        self.out_features=out_features
        self.weights=Parameter(torch.Tensor(out_features, in_features))
        if bias:
            self.bias=Parameter(torch.Tensor(out_features))
        else:
            self.register_parameter('bias', None)
        self.reset_parameters()
    def forward(self, input):
        return F.linear(input, self.weight, self.bias)
        
```

* Diffrence betwwen Parameter and Argument
    1. Parameters are used inside fucntion defination, Placeholder, Actual value passed to thr function when function is called. They store some data.
        1. Hyper Parameters
        2. Data Dependenet Parameter
    2. Arguments, come from outside and are used to assign the values to the function
    
##### A. Manually choosen Hyper Parameter
* This are the one, NN architet has to decide on his own
    1. Kernel Size : FIlter Size
    2. Out Channel: Number of filter, one filter = one channel
    3. Out Features: Size of output teensor, feature map same as Out channel

##### B. Data Dependent Parameter
* Values depend on data
    1. In_channel of 1st Conv Layer =1. Start 
    2. Out_Features of last Linear layer=10. End

#### C. Learnable Parameters
* Whose values are learned during the training process. Initalizae randomly at start.

## CNN Forward Method

### Input Layer
* f(x)=x or identity transformation as argument is asme as input

```python
class Network(nn.Module): #Line 1
    def __init__(self):
        super(Network, self).__init__() # Line 3
        
        self.conv1=nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5)
        self.conv2=nn.Conv2d(in_channels=6, out_channels=12, kernel_size=5)
        
        self.fc1=nn.Linear(in_features=12*4*4, out_features=120)# 12 comnes from output of 12 channel in conv2
        self.fc2=nn.Linear(in_features=120, out_features=60)
        
        self.out=nn.Linear(in_features=60, out_features=10)
        
    def forward(self, t):
        # (1) Input Layer
        t=t
        
        # (2) hidden conv1 Layer
        t=self.conv1(t)
        t=F.relu(t)
        t=F.max_pool2d(t, kernel_size=2, stride=2)
        
        # (3) hidden conv2 Layer
        t=self.conv2(t)
        t=F.relu(t)
        t=F.max_pool2d(t, kernel_size=2, stride=2)
        
        # (4) hidden linear1 layer
        t=t.reshape(-1, 12*4*4) # Need to flatten before using FC layer 
        t=self.fc1(t)
        t=F.relu(t)
        
        # (5) Hidden linear2 layer
        t=self.fc2(t)
        t=F.relu(t)
        
        # (6) Output layer
        t=self.out(t)
        # Avoiding softmax because we will use cross entropy loss fucntion which has built in softmax fucntion.
        #t=F.softmax(t, dim=1) 
        # THhis alllows the network to not compute the additiional operation while network is been infrenced. But computaion will of course work.
        
        
        
        return t
```

## Training Process

1. Get batch from training set
2. Pass batch to network
3. Calculate loss
4. Calculate Gradient
5. Update Weights
6. Repeat 1-5 till one epoch
7. Repeat 1-6 till as many epoch needed to obtain accuracy

    * Epoch means one pass of data

```python
torch.set_grad_enabled(True) # In past we turned the off. THis time though we will turn them on
```

```python
def accuracy_nums(preds,labels):
    return (preds.argmax(dim=1).eq(labels).sum().item())
```

```python
class Network(nn.Module): #Line 1
    def __init__(self):
        super().__init__() # Line 3
        
        self.conv1=nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5)
        self.conv2=nn.Conv2d(in_channels=6, out_channels=12, kernel_size=5)
        
        self.fc1=nn.Linear(in_features=12*4*4, out_features=120)# 12 comnes from output of 12 channel in conv2
        self.fc2=nn.Linear(in_features=120, out_features=60)
        
        self.out=nn.Linear(in_features=60, out_features=10)
        
    def forward(self, t):
        # (1) Input Layer
        #t=t
        
        # (2) hidden conv1 Layer
        t=self.conv1(t)
        t=F.relu(t)
        t=F.max_pool2d(t, kernel_size=2, stride=2)
        
        # (3) hidden conv2 Layer
        t=self.conv2(t)
        t=F.relu(t)
        t=F.max_pool2d(t, kernel_size=2, stride=2)
        
        # (4) hidden linear1 layer
        t=t.reshape(-1, 12*4*4) # Need to flatten before using FC layer 
        t=self.fc1(t)
        t=F.relu(t)
        
        # (5) Hidden linear2 layer
        t=self.fc2(t)
        t=F.relu(t)
        
        # (6) Output layer
        t=self.out(t)
        # Avoiding softmax because we will use cross entropy loss fucntion which has built in softmax fucntion.
        #t=F.softmax(t, dim=1) 
        # THhis alllows the network to not compute the additiional operation while network is been infrenced. But computaion will of course work.
        
        
        
        return t
network=Network()
```

* All the above can be encapsulated as follows:

```python
train_loader=torch.utils.data.DataLoader(
    train_set, batch_size=100 #Default batach size =1
)
optimizer=optim.Adam(network.parameters(), lr=0.01)


batch=next(iter(train_loader))# Get Batch
images, labels =batch

preds=network(images) # Predict from Batch
loss=F.cross_entropy(preds, labels)# This has the all the data to for back prop

loss.backward() # Back-prop to find gradients
optimizer.step() # Updating weights

#-------------------------------------------------------------
print("loss1:", loss.item())
preds=network(images)
loss=F.cross_entropy(preds,labels)

print("loss2:", loss.item())
preds=network(images)
loss=F.cross_entropy(preds,labels)
```

#### 6. Training Loop

```python
train_loader=torch.utils.data.DataLoader(
    train_set, batch_size=100 #Default batach size =1
)
optimizer=optim.Adam(network.parameters(), lr=0.01)

for epoch in range(5):
    total_loss=0
    total_correct=0
    for batch in train_loader:
        images, labels =batch

        preds=network(images) # Predict from Batch
        loss=F.cross_entropy(preds, labels)# This has the all the data to for back prop

        optimizer.zero_grad() # Needed becuase Pytorch makes gradiesnt cumulative, which we do not want
        loss.backward() # Back-prop to find gradients
        optimizer.step() # Updating weights
        total_correct+=accuracy_nums(preds,labels)    


        total_loss+=loss.item()
    print("Epoch,",epoch,"Total Correct:", total_correct, "loss:", total_loss)    
```

```python
acr=(total_correct/len(train_set)*100)
print("Accuracy %.2f%%" %acr )
```

##### Getting predictions for entire training set

```python
def get_all_preds(model,loader):
    all_preds=torch.tensor([])
    
    for batch in loader:
        images, labels =batch
        preds=model(images)
        all_preds=torch.cat(
                        (all_preds, preds),
                        dim=0
                        )
    
    return all_preds
```
